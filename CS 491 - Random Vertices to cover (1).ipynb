```python
from matplotlib import pyplot as plt
import random
import math
%matplotlib inline
import pandas as pd
import csv
```

```python
with open('600CellFacesMadeWithCoordinates.csv', 'rt') as f:
    reader = csv.reader(f)
    facesMatrix1 = list(reader)
```

```python
def solChecker(vertArr, faceMatrix):
    #loops through each face
    for face in faceMatrix:
        #makes a list comperhension of all the verts in vert array that it is toching 
        #then checks if the length is 0
        if len([i for i in vertArr if i in face]) == 0:
            #if it is then it returns false
            return False
    #if it runs through every face then it is true
    return True
```

```python
runs = 50000
sixhundredCellSim = []
for i in range(40):
    sixhundredCellSim.append(0)

for num in range(40, 121):
    solutions = 0
    for i in range(runs):
        randomVertsInt = random.sample(range(1,121),num)
        randomVerts = list(map(str, randomVertsInt))
        ##print(randomVerts)
        if(solChecker(randomVerts,facesMatrix1)):
            solutions += 1
    sixhundredCellSim.append((solutions/runs)*100)
```

```python
a= plt.figure()

a.set_figwidth(20)
a.set_figheight(10)

axes= a.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlim([40,120])
plt.plot(sixhundredCellSim, 'b')

plt.axhline(y=50, color='r', linestyle='-')
plt.ylabel('Percent of solutions Correct')
plt.xlabel('Amount of Vertices')
plt.show()
```

```python
df = pd.DataFrame(sixhundredCellSim, columns=['% of Random Solutions Correct'])
pd.set_option('display.max_rows', None)
df = df.iloc[40:]

df
```
